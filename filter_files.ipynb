```python
import os
import re

# Define the folder path
folder_path = 'processed_files'

# Define the regex patterns
pattern_dates = r'\(20(?:1|2)\d\)'
pattern_spanish = r'dos mil(?: |\n)(?:once|doce|trece|catorce|quince|dieciseis|dieciséis|diecisiete|dieciocho|diecinueve|veinte|veintiuno|veitiuno|veintidós|veintitrés|veintidos|veintitres|venticuatro|dlecisiete)'
pattern_alternative_dates = r'20(?:1|2)\d'

# Initialize dictionaries and lists
results = {}
no_matches = []
spanish_matches = {}
no_spanish_matches = []
alternative_matches = {}
no_alternative_matches = []

# Process all files
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)
        # Open and read the file
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read().lower()  # Convert text to lowercase
            # Find all matches for the primary date pattern
            matches = re.findall(pattern_dates, text)
            if matches:
                # Store the first match if there are multiple
                results[filename] = matches[0]
            else:
                # Add to the no-match list if no matches found
                no_matches.append(filename)

# Process files with no initial matches for Spanish word dates
for filename in no_matches:
    file_path = os.path.join(folder_path, filename)
    # Open and read the file
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().lower()  # Convert text to lowercase
        # Find all matches for the Spanish date pattern
        spanish_matches_found = re.findall(pattern_spanish, text)
        if spanish_matches_found:
            spanish_matches[filename] = spanish_matches_found[0]
        else:
            # Add to the no-spanish-match list if no Spanish date matches found
            no_spanish_matches.append(filename)

# Process files with no Spanish matches for alternative dates
for filename in no_spanish_matches:
    file_path = os.path.join(folder_path, filename)
    # Open and read the file
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read().lower()  # Convert text to lowercase
        # Find all matches for the alternative date pattern
        alternative_matches_found = re.findall(pattern_alternative_dates, text)
        if alternative_matches_found:
            alternative_matches[filename] = alternative_matches_found[0]
        else:
            # Add to the no-alternative-match list if no alternative date matches found
            no_alternative_matches.append(filename)

# Count how many extracted dates are after 2019


# Print the results
print("Extracted dates:", results)
print("Files with no match:", no_matches)
print("Spanish matches:", spanish_matches)
print("Files with no Spanish match:", no_spanish_matches)
print("Alternative matches:", alternative_matches)
print("Files with no alternative match:", no_alternative_matches)
```

```python
year_pattern = r' 20(?:1|2)\d'

# New dictionary to store the extracted years
extracted_years_name = {}

# List to store keys with no year matches
no_year_matches = []

# Extract keys and match year using regex
for key in no_spanish_matches:
    match = re.search(year_pattern, key)
    if match:
        extracted_years_name[key] = int(match.group().strip())
    else:
        no_year_matches.append(key)

print("Extracted Years Dictionary:")
print(extracted_years_name)
print("Keys with No Year Matches:")
print(no_year_matches)
print(len(no_year_matches))
```

```python
import re

def convert_spanish_dates(spanish_dict):
    # Mapping Spanish words to years
    spanish_to_year = {
        'once': '11', 'doce': '12', 'trece': '13', 'catorce': '14', 'quince': '15',
        'dieciseis': '16', 'dieciséis': '16', 'diecisiete': '17', 'dlecisiete':'17','dieciocho': '18',
        'diecinueve': '19', 'veinte': '20', 'veintiuno': '21', 'veintiuno':'21', 'veintidós': '22',
        'veintitrés': '23', 'veintidos': '22', 'veintitres': '23', 'venticuatro': '24',
        'dlecisiete': '17'
    }
    
    # Function to convert Spanish date to a numeric year
    def spanish_to_numeric(spanish_date):
        # print(spanish_date)
        
        spanish_year = spanish_date.replace('\n', ' ').split(' ')[-1]
        year_suffix = spanish_to_year.get(spanish_year, '')
        return f"20{year_suffix}"
    
    # Create a new dictionary with converted years
    converted_dict = {}
    for filename, spanish_date in spanish_dict.items():
        # Convert each Spanish date to numeric format
        numeric_date = int(spanish_to_numeric(spanish_date))
        if numeric_date:
            converted_dict[filename] = numeric_date
    
    return converted_dict


converted_spanish_matches = convert_spanish_dates(spanish_matches)
print(converted_spanish_matches)
```

```python
def clean_date_dict(date_dict):
    # Function to clean the date string
    def clean_date(date_str):
        # Remove parentheses and convert to integer
        date_str = date_str.strip('()')  # Remove parentheses
        return int(date_str) if date_str.isdigit() else None
    
    # Create a new dictionary with cleaned dates
    cleaned_dict = {}
    for filename, date_str in date_dict.items():
        cleaned_date = clean_date(date_str)
        if cleaned_date is not None:
            cleaned_dict[filename] = cleaned_date
    
    return cleaned_dict



cleaned_dates_pattern_1 = clean_date_dict(results)
print(cleaned_dates_pattern_1)
```

```python
def convert_dates_to_int(date_dict):
    # Function to convert date string to integer
    def to_int(date_str):
        # Remove any extraneous characters and convert to integer
        return int(date_str) if date_str.isdigit() else None
    
    # Create a new dictionary with integer values
    int_dict = {}
    for filename, date_str in date_dict.items():
        # Convert date string to integer
        int_date = to_int(date_str)
        if int_date is not None:
            int_dict[filename] = int_date
    
    return int_dict


cleaned_dates_pattern_2 = convert_dates_to_int(alternative_matches)
print(cleaned_dates_pattern_2)
```

```python
import os

def count_files_in_directory(directory):
    # Get a list of all files in the directory
    files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
    # Count the number of files
    return len(files)

# Define the folder path
folder_path = 'processed_files'

# Count files in the directory
file_count = count_files_in_directory(folder_path)

print(f"Number of files in '{folder_path}': {file_count}")
```
